## 1.‚ö° Concepto de Programaci√≥n Orientada a Objetos ‚ö°

* La programaci√≥n orientada a objetos (POO) es un paradigma de programaci√≥n que **se enfoca en la creaci√≥n de "objetos" que contienen tanto datos como funcionalidad.** 


* Estos objetos est√°n dise√±ados para **interactuar entre s√≠ y trabajar juntos para resolver problemas complejos.** 


* En Python, **todo es un objeto, incluyendo los tipos de datos b√°sicos como las cadenas y los n√∫meros.**


* Estos objetos se relacionan entre s√≠ y se utilizan para **representar conceptos del mundo real en el c√≥digo.**     

    * Un estudiante
    * Un coche
    * Un tel√©fono
    * Una persona
    * etc 


* Primer pilar de la POO: **los datos y las operaciones que pueden realizarse** con esos datos se agrupan en unidades l√≥gicas llamadas:

 ‚≠ê **OBJETOS.** ‚≠ê 


Los otros dos pilares de la POO son la **herencia y el polimorfismo** que veremos m√°s adelante.


<font color='blue'>**Las principales ventajas de la POO son:**</font>

1. **Facilita la organizaci√≥n del c√≥digo de una manera m√°s clara y l√≥gica,** lo que hace que sea m√°s legible y comprensible.


2. Reduce la cantidad de c√≥digo que se tiene que escribir y **facilita el mantenimiento del programa.**


3. La POO facilita el trabajo con colecciones de datos, ya que **permite tratar a cada objeto de manera individual.**


4. Permite hacer una mejor distinci√≥n entre la l√≥gica de negocio y la interfaz de usuario, lo que **facilita el desarrollo de programas complejos.**


5. **Facilita el trabajo en equipo,** ya que cada miembro del equipo puede trabajar en una parte diferente del programa sin afectar el trabajo de los dem√°s.

### 2.4 Constructores

* El constructor es un **m√©todo especial en una clase que se llama autom√°ticamente cuando se crea un objeto de esa clase.**


* El constructor se llama "\__init__" y se utiliza para inicializar los atributos del objeto reci√©n creado. 


* El constructor **siempre debe tener al menos un argumento, que se llama "self"** y que hace referencia al objeto 
que est√° siendo creado. Por ejemplo:

```python
class MiClase:
    def __init__(self):
        # Inicializar atributos del objeto
        self.mi_atributo = 5
```

#### 2.4.1 Definiendo Varios Constructores.

**¬øPuedo tener varios m√©todos para crear un objeto?**

* S√≠, en Python **se pueden tener varios constructores en una misma clase.** 


* Esto se logra mediante la sobrecarga de m√©todos, que **permite tener varias implementaciones del mismo m√©todo con diferentes argumentos.**


* Cuando se quiere tener varios constructores, **se suele tener un constructor "principal" que tiene todos los argumentos necesarios para inicializar un objeto,** y luego otros constructores que llaman al constructor principal y le pasan valores por defecto para algunos de sus argumentos. De esta manera, se pueden proporcionar diferentes formas de crear un objeto de la clase. Por ejemplo:

```python
class MiClase:
    def __init__(self, arg1, arg2, arg3):
        # Constructor principal
        # Inicializar atributos con los argumentos recibidos
        self.arg1 = arg1
        self.arg2 = arg2
        self.arg3 = arg3

    def __init__(self, arg1, arg2):
        # Constructor secundario que llama al constructor principal
        # Se asume que arg3 tiene un valor por defecto
        self.__init__(arg1, arg2, "Valor por defecto")

    def __init__(self):
        # Constructor secundario que llama al constructor principal
        # Se asumen valores por defecto para todos los argumentos
        self.__init__("Valor por defecto", "Valor por defecto", "Valor por defecto")
```

En este ejemplo, **se pueden crear objetos de la clase "MiClase" de tres maneras diferentes:** proporcionando todos los argumentos al constructor, proporcionando solo dos argumentos, o no proporcionando argumentos (en cuyo caso se usar√°n los valores por defecto).

## 3. Creando Objetos - üîß**Instanciaci√≥n de clases**üîß 


* Llamamos instanciar a la acci√≥n de crear un objeto de una determinada clase.

* Para crear o instanciar un objeto de una determinada clase se utiliza:

  * **El nombre de la clase seguida de los par√°metros necesarios para crear el objeto entre par√©ntesis.**


* Cuando se crea un objeto de una clase se dice que el objeto ‚òù**es una instancia de la clase.**‚òù


* Podemos tener **infinitos objetos instanciados** de una determinada clase.

```python
class MiClase:
    def __init__(self, arg1, arg2):
        # Inicializar atributos del objeto
        print("Creando objeto MiClase")
        self.arg1 = arg1
        self.arg2 = arg2
        
    def mi_metodo(self):
        print("Llamando a Mi metodo")
```

```python
# Crear un objeto de la clase "MiClase"
objeto = MiClase("Valor 1", "Valor 2")
```

**Una vez creado el objeto, se pueden acceder a sus atributos y utilizar sus m√©todos mediante la notaci√≥n "objeto.atributo" o "objeto.metodo()".** 

Por ejemplo:

```python
# Acceder a un atributo del objeto
valor_atributo = objeto.arg1
print(valor_atributo)

# Llamar a un m√©todo del objeto
objeto.mi_metodo()
```

## 4. M√©todos M√°gicos

* Los m√©todos m√°gicos **son m√©todos especiales que tienen una sintaxis con doble gui√≥n bajo al principio y al final de su nombre.** Estos m√©todos tienen un **comportamiento especial en Python** y se utilizan para implementar ciertas funcionalidades que no pueden ser realizadas mediante el uso de m√©todos normales.

* Por ejemplo, el m√©todo m√°gico \__init__ se utiliza para inicializar una clase cuando se crea una nueva instancia de esa clase. Este m√©todo **se ejecuta autom√°ticamente cuando se crea una nueva instancia de una clase** y se utiliza para establecer el estado inicial de la nueva instancia.



### 4.1 Definiendo M√©todos M√°gicos

* Para definir un m√©todo m√°gico en Python, se utiliza la sintaxis def __nombre_del_m√©todo__(self):. 

* El primer y √∫ltimo doble gui√≥n bajo es obligatorio y el nombre del m√©todo debe estar entre estos dos guiones bajos. 

* Adem√°s, todos los m√©todos m√°gicos **deben incluir el argumento self**, que se refiere a la instancia de la clase en la que se est√° ejecutando el m√©todo.

Para utilizar un m√©todo m√°gico, simplemente **se lo llama como cualquier otro m√©todo,** pero utilizando la sintaxis especial que **incluye los dobles guiones** bajos al principio y al final del nombre del m√©todo.

Algunos de estos m√©todos **se llaman de forma autom√°tica** cuando creamos un objeto o utilizamos alguna funci√≥n de python.

### 4.2 STR - Imprimiendo Objetos

* Otro ejemplo de un m√©todo m√°gico es el m√©todo **\__str__**, que se **utiliza para representar una clase como una cadena de caracteres.** 

* Este m√©todo se llama autom√°ticamente cuando **se intenta convertir una instancia de una clase en una cadena de caracteres,** por ejemplo mediante la llamada a la funci√≥n str() o cuando se imprime una instancia de una clase utilizando la funci√≥n print().

### 4.3 Otros M√©todos M√°gicos

Hay muchos otros m√©todos m√°gicos en Python, cada uno con una funci√≥n espec√≠fica. Algunos ejemplos de m√©todos m√°gicos son:

* **\__add__:** se utiliza para sobrecargar el operador + y determinar c√≥mo se sumar√°n dos instancias de una clase.

* **\__len__:** se utiliza para determinar cu√°l es la longitud de una instancia de una clase. 
  * Por ejemplo, si se tiene una lista implementada mediante una clase y **se llama a la funci√≥n len() sobre una instancia de esta clase,** se ejecutar√° el m√©todo **\__len__:** para determinar cu√°ntos elementos hay en la lista.


* **\__contains__:** se utiliza para determinar si una instancia de una clase **contiene un determinado elemento.** 

  * Por ejemplo, si se tiene una lista implementada mediante una clase y **se llama al operador in sobre una instancia de esta clase**, se ejecutar√° el m√©todo **\__contains__: para determinar si la lista contiene el elemento buscado.


Hay muchos otros m√©todos m√°gicos en Python y cada uno de ellos se utiliza para implementar una funcionalidad espec√≠fica. Para conocer todos los m√©todos m√°gicos disponibles en Python y su uso, se recomienda consultar la [documentaci√≥n oficial de Python](https://docs.python.org/3/)
